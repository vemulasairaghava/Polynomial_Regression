```python
from google.colab import drive
import pandas as pd


drive.mount('/content/drive')


path = '/content/drive/My Drive/Position_Salaries.csv'
```

```python
dataset=pd.read_csv(path)
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

```

```python
# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

print("✅ Models Trained Successfully!")
```

```python
import matplotlib.pyplot as plt
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
linear_prediction = lin_reg.predict([[6.5]])
print("Linear prediction: ",linear_prediction)
polynomial_prediction = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print("Polynomial prediction: ",polynomial_prediction)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
import numpy as np
mse=mean_squared_error(y,lin_reg.predict(X))
rmse=np.sqrt(mse)
mae=mean_absolute_error(y,lin_reg.predict(X))
mape=mean_absolute_percentage_error(y,lin_reg.predict(X))
r2=r2_score(y,lin_reg.predict(X))
print("MSE: ",mse)
print("RMSE: ",rmse)
print("MAE: ",mae)
print("MAPE: ",mape)
print("R2: ",r2)
```

```python
mse1=mean_squared_error(y,lin_reg_2.predict(poly_reg.fit_transform(X)))
rmse1=np.sqrt(mse1)
mae1=mean_absolute_error(y,lin_reg_2.predict(poly_reg.fit_transform(X)))
mape1=mean_absolute_percentage_error(y,lin_reg_2.predict(poly_reg.fit_transform(X)))
r21=r2_score(y,lin_reg_2.predict(poly_reg.fit_transform(X)))
print("Mse for polynomial: ",mse1)
print("Rmse for polynomial: ",rmse1)
print("Mae for polynomial: ",mae1)
print("Mpe for polynomial: ",mape1)
print("R2 for polynomial: ",r21)
```
